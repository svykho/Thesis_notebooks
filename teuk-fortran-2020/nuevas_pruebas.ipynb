## lin_f_m2_re.csv

```python
# CSV del output
file_path = '/home/alex/Desktop/tfm-xisco/teuk-fortran-2020/output/Feb__4_m0.5_s0.35_nx48_nl24/lin_f_m2_re.csv' 
data = pd.read_csv(file_path)

# Separamos las columnas para distinguir entre tiempo y espacio
time = pd.to_numeric(data.iloc[:, 0], errors='coerce')  # Primera columna: tiempo
psi_values_horizon = pd.to_numeric(data.iloc[:, 3], errors='coerce')  # Horizonte (primer punto espacial)
psi_values_infinity = pd.to_numeric(data.iloc[:, -2], errors='coerce')  # Infinito nulo futuro (Ãºltimo punto espacial)
```

```python
fig, axes = plt.subplots(1, 2, figsize=(14, 6))

# Horizonte (lineal)
axes[0].plot(time, psi_values_horizon, label=r"$R\Psi_4^{(1)[2]}$ (Horizon)", color="fuchsia")
axes[0].set_xlabel(r"$t/M$", fontsize=14)
axes[0].set_ylabel(r"$R\Psi_4^{(1)[2]}$", fontsize=14)
axes[0].set_title(r"Real part of $\Psi_4^{(1)}$ at Horizon ($m=2$)", fontsize=16)
axes[0].legend(fontsize=12)
axes[0].grid()

# Horizonte (logarÃ­tmica)
axes[1].semilogy(time, abs(psi_values_horizon), label=r"$|R\Psi_4^{(1)[2]}|$ (Horizon)", color="fuchsia")
axes[1].set_xlabel(r"$t/M$", fontsize=14)
axes[1].set_ylabel(r"$|R\Psi_4^{(1)[2]}|$", fontsize=14)
axes[1].set_title(r"Real part of $\Psi_4^{(1)}$ at Horizon ($m=2$) [Log Scale]", fontsize=16)
axes[1].legend(fontsize=12)
axes[1].grid()

plt.tight_layout()
plt.show()
```

```python
fig, axes = plt.subplots(1, 2, figsize=(14, 6))

# Infinito nulo futuro (lineal)
axes[0].plot(time, psi_values_infinity, label=r"$R\Psi_4^{(1)[2]}$ (Future Null Infinity)", color="blue")
axes[0].set_xlabel(r"$t/M$", fontsize=14)
axes[0].set_ylabel(r"$R\Psi_4^{(1)[2]}$", fontsize=14)
axes[0].set_title(r"Real part of $\Psi_4^{(1)}$ at Future Null Infinity ($m=2$)", fontsize=16)
axes[0].legend(fontsize=12)
axes[0].grid()

# Infinito nulo futuro (logarÃ­tmica)
axes[1].semilogy(time, abs(psi_values_infinity), label=r"$|R\Psi_4^{(1)[2]}|$ (Future Null Infinity)", color="blue")
axes[1].set_xlabel(r"$t/M$", fontsize=14)
axes[1].set_ylabel(r"$|R\Psi_4^{(1)[2]}|$", fontsize=14)
axes[1].set_title(r"Real part of $\Psi_4^{(1)}$ at Future Null Infinity ($m=2$) [Log Scale]", fontsize=16)
axes[1].legend(fontsize=12)
axes[1].grid()

plt.tight_layout()
plt.show()
```
