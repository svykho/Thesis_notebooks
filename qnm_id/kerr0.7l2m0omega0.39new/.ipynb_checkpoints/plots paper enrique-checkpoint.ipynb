## FIGURE 1

```python
# Definimos la expresin completa de |(k)|^2 de la ecuacin (2.66)

def xi_squared_full(kdx, alpha):
    sin_kdx2 = np.sin(kdx / 2)
    cos_kdx = np.cos(kdx)
    cos_2kdx = np.cos(2 * kdx)
    cos_3kdx = np.cos(3 * kdx)
    
    term1 = 1
    term2 = (8/9) * alpha**2 * (cos_kdx - 5) * sin_kdx2**6
    term3 = -(4/9) * alpha**4 * (4 * cos_kdx - 17) * sin_kdx2**8
    term4 = (1/54) * alpha**6 * (133 * cos_kdx - 34 * cos_2kdx + 3 * cos_3kdx - 150) * sin_kdx2**6
    term5 = (4/81) * alpha**8 * (cos_kdx - 4)**2 * sin_kdx2**8

    return term1 + term2 + term3 + term4 + term5

# Rango de k * dx
kdx = np.linspace(0, 2*np.pi, 500)
alphas = [0.25, 0.5, 1.0, 1.5, 1.6]

# Graficar
plt.figure(figsize=(9, 6))
for alpha in alphas:
    xi_sq = xi_squared_full(kdx, alpha)
    plt.plot(kdx, xi_sq, label=f"{alpha:.2f}")

plt.xlabel(r"$k \, \delta x$")
plt.ylabel(r"$|\xi(k)|^2$")
plt.title(r"Factor de amplificacin $|\xi(k)|^2$ para el esquema de cuarto orden")
plt.legend(title=r"$\alpha$")
plt.grid(True)
plt.tight_layout()
plt.show()
```
